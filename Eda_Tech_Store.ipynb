#Importações e Carregamento dos Dados


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df_dados = pd.read_csv("https://raw.githubusercontent.com/GustavoMillerOS/Eda_Tech_Store/refs/heads/main/vendas.csv")
```

```
# Converter coluna Data para tipo datetime
df_dados['data'] = pd.to_datetime(df_dados['data'])
```

```
# Calcular a receita total
df_dados["Receita_Total"] = df_dados['quantidade'] * df_dados['preco_unitario']
print(df_dados[['quantidade', 'preco_unitario', 'Receita_Total']].head())
```

```
# Filtrar e exibir as vendas da categoria "Eletrônicos".
vendas_eletronicos = df_dados[df_dados['categoria'] == 'Eletrônicos']
vendas_eletronicos
```

```
# Identificar e exibir o produto mais vendido (em quantidade).
vendas_por_produto = df_dados.groupby('produto')['quantidade'].sum()
produto_mais_vendido = vendas_por_produto.sort_values(ascending=False).idxmax()
quantidade_maxima = vendas_por_produto.max()

print(f"O produto mais vendido é: {produto_mais_vendido}")
print(f"Total de unidades vendidas: {quantidade_maxima}")
```

```
# Descobrir e exibir a região com maior valor de compras.
vendas_regiao = df_dados.groupby('regiao')['Receita_Total'].sum()
regiao_lider = vendas_regiao.idxmax()
valor_maximo = vendas_regiao.max()

print(f"A região com maior valor de compras é: {regiao_lider}")
print(f"Valor total: R$ {valor_maximo:,.2f}")
```

```
# Gráfico de barras mostrando a receita por categoria.
receita_categoria = df_dados.groupby('categoria')['Receita_Total'].sum().sort_values(ascending=False)

receita_categoria.plot(kind='bar', color='skyblue', figsize=(10,5))
plt.title('Receita Total por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Receita(R$)')
plt.xticks(rotation=45)
plt.show()
```

```
# Gráfico de linha mostrando a evolução das vendas por mês.
evolucao_mensal = df_dados.groupby(df_dados['data'].dt.to_period('M'))['Receita_Total'].sum()
evolucao_mensal.index = evolucao_mensal.index.astype(str)
evolucao_mensal.plot(kind='line', marker='o', color='green', figsize=(10,5))
plt.title('Evolução Mensal de Vendas')
plt.xlabel('Mês')
plt.ylabel('Receita (R$)')
plt.grid(True)
plt.show()
```

```
# Montar uma tabela dinâmica com receita por região × categoria.
tabela_dinamica = df_dados.pivot_table(index='regiao',
                                       columns='categoria',
                                       values='Receita_Total',
                                       aggfunc='sum',
                                       fill_value=0)
tabela_dinamica
```

```
# Criar um relatório exportado em .xlsx
with pd.ExcelWriter('Relatorio_Vendas.xlsx') as writer:
    df_dados.to_excel(writer, sheet_name='Dados_Gerais', index=False)
    tabela_dinamica.to_excel(writer, sheet_name='Resumo_Regiao_Categoria')

print("Relatório exportado com sucesso! Procure por 'Relatorio_Vendas.xlsx' na barra lateral.")
```
